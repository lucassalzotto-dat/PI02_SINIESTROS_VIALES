```python
# Configurar estilo
sns.set(style="ticks")

# Crear subgráficos
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Subplot para todos los datos
sns.countplot(x='sexo_victima', data= victimas[victimas['gravedad'] != 'fatal'].dropna(subset=['sexo_victima']), ax=axs[0])
axs[0].set_title('Distribución de Sexo de la Víctima',fontsize=16)
axs[0].set_xlabel('')
axs[0].set_ylabel('Frecuencia')

# Agregar etiquetas de porcentaje en cada barra (primer subplot)
total_data = len(victimas[victimas['gravedad'] != 'fatal'].dropna(subset=['sexo_victima']))
for p in axs[0].patches:
    height = p.get_height()
    axs[0].text(p.get_x() + p.get_width() / 2., height + 0.1, f'{height/total_data:.2%}', ha="center")

# Subplot solo para datos de gravedad fatal
sns.countplot(x='sexo_victima', data=victimas[victimas['gravedad'] == 'fatal'].dropna(subset=['sexo_victima']), ax=axs[1])
axs[1].set_title('Distribución de Sexo de la Víctima (Fatales)',fontsize= 16)
axs[1].set_xlabel('')
axs[1].set_ylabel('Frecuencia')

# Agregar etiquetas de porcentaje en cada barra (segundo subplot)
total_fatal = len(victimas[victimas['gravedad'] == 'fatal'].dropna(subset=['sexo_victima']))
for p in axs[1].patches:
    height = p.get_height()
    axs[1].text(p.get_x() + p.get_width() / 2., height + 0.1, f'{height/total_fatal:.2%}', ha="center")

# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()
plt.show()
```

Conclusión: La preponderancia de víctimas, tanto en accidentes no fatales como fatales, recae mayoritariamente en el género masculino.

```python
# Cargar el archivo Shapefile de las comunas
shapefile = gpd.read_file("../Datasets/Archivos_mapa/comunas_wgs84.shp")

# Filtrar puntos dentro del rango de las comunas
hechos_filtered = hechos[(hechos['longitud'] >= shapefile.bounds['minx'].min()) & 
                          (hechos['longitud'] <= shapefile.bounds['maxx'].max()) &
                          (hechos['latitud'] >= shapefile.bounds['miny'].min()) & 
                          (hechos['latitud'] <= shapefile.bounds['maxy'].max())]

# Filtrar los datos según la columna 'gravedad'
hechos_fatal = hechos_filtered[hechos_filtered['gravedad'] == 'fatal']
hechos_no_fatal = hechos_filtered[hechos_filtered['gravedad'] != 'fatal']

# Iniciar CRS
crs = {'init': 'epsg:4326'}

# Definir la geometría para cada conjunto de datos
geometry_fatal = [Point(xy) for xy in zip(hechos_fatal["longitud"], hechos_fatal["latitud"])]
geometry_no_fatal = [Point(xy) for xy in zip(hechos_no_fatal["longitud"], hechos_no_fatal["latitud"])]

# Crear GeoDataFrames para cada conjunto de datos
geodata_fatal = gpd.GeoDataFrame(hechos_fatal, crs=crs, geometry=geometry_fatal)
geodata_no_fatal = gpd.GeoDataFrame(hechos_no_fatal, crs=crs, geometry=geometry_no_fatal)

# Crear ejes y figura
fig, ax = plt.subplots(figsize=(10, 8))

# Dibujar el Shapefile
shapefile.plot(ax=ax, edgecolor='k', alpha=0.3, linewidth=2,cmap="inferno")

# Dibujar los datos geoespaciales para gravedad 'fatal' (en rojo)
geodata_fatal.plot(ax=ax, color='red', markersize=14, alpha=1, label='Fatal')

# Dibujar los datos geoespaciales para gravedad distinta de 'fatal' (en azul)
geodata_no_fatal.plot(ax=ax, color='green', markersize=7, alpha=0.1, label='No fatal')

# Ajustar la relación de aspecto manualmente y desactivar la configuración automática
ax.set_aspect('equal', adjustable='datalim')

# Añadir espacio superior adicional para el título
fig.subplots_adjust(top=0.95)

# Agregar título y etiquetas de ejes
fig.suptitle('Mapa de accidentes', fontsize=16)
ax.set_xlabel('', fontsize=10)
ax.set_ylabel('', fontsize=10)

# Ajustar la leyenda
legend = ax.legend()
legend.legendHandles[1].set_alpha(1.0)
legend.legendHandles[1]._sizes = [14]

# Mostrar el mapa
plt.show()
```

En el mapa se pueden apreciar los incidentes de tráfico en la Ciudad de Buenos Aires. Los puntos verdes señalan las ubicaciones donde se han registrado accidentes no fatales, mientras que los puntos rojos representan los incidentes fatales.

```python

# Crear subconjuntos de datos
todos_los_datos = victimas
gravedad_fatal = victimas[victimas['gravedad'] == 'fatal']

# Configurar estilo
sns.set(style="ticks")


# Crear subgráficos 2x2 con anchos relativos
fig, axs = plt.subplots(2, 2, figsize=(15, 10), gridspec_kw={'width_ratios': [8, 2]})

# Histograma de todas las edades
sns.histplot(data=todos_los_datos, x='edad_victima', ax=axs[0, 0],bins= 12)
axs[0, 0].set_title('Histograma de Edades',fontsize=18)
axs[0, 0].set_xlabel('Edad')
axs[0, 0].set_ylabel('Frecuencia')

# Boxplot de todas las edades (vertical)
sns.boxplot(data=todos_los_datos, y='edad_victima', ax=axs[0, 1])
axs[0, 1].set_title('Boxplot de Edades',fontsize=18)
axs[0, 1].set_xlabel('Edad')
axs[0, 1].set_ylabel('')

# Histograma de edades con gravedad fatal
sns.histplot(data=gravedad_fatal, x='edad_victima', ax=axs[1, 0],bins= 12)
axs[1, 0].set_title('Histograma de Edades (Gravedad Fatal)',fontsize=18)
axs[1, 0].set_xlabel('Edad')
axs[1, 0].set_ylabel('Frecuencia')

# Boxplot de edades con gravedad fatal (vertical)
sns.boxplot(data=gravedad_fatal, y='edad_victima', ax=axs[1, 1])
axs[1, 1].set_title('Boxplot de Edades (Gravedad Fatal)',fontsize=18)
axs[1, 1].set_xlabel('Edad')
axs[1, 1].set_ylabel('')

# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()
plt.show()
```

Conclusión: Al analizar detenidamente los histogramas que representan las edades en accidentes, tanto fatales como no fatales, observamos una sorprendente similitud en la forma de sus distribuciones. No obstante, al explorar el boxplot que abarca todos los datos, resalta la presencia de valores extremos que no se encuentran presentes en los accidentes de gravedad fatal. Aunque las medianas en ambos escenarios son parecidas, indicando una tendencia central similar, es destacable que el rango intercuartílico de los accidentes fatales es notablemente más amplio en comparación con los no fatales. Esto sugiere una mayor variabilidad en las edades de las víctimas en situaciones de gravedad fatal.

```python
# Filtrar gravedades distintas de 'fatal' y 'fatal'
gravedad_no_fatal = hechos[hechos['gravedad'] != 'fatal']
gravedad_fatal = hechos[hechos['gravedad'] == 'fatal']

# Combina las dos columnas en una sola serie y elimina los valores nulos
vehiculos_no_fatal = pd.concat([gravedad_no_fatal['vehiculo1'], gravedad_no_fatal['vehiculo2']]).dropna()
vehiculos_fatal = pd.concat([gravedad_fatal['vehiculo1'], gravedad_fatal['vehiculo2']]).dropna()

# Configuración del estilo de Seaborn 
sns.set(style="ticks")

# Crear subgráficos 
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(16, 12))

# Subgráfico 1: Gravedades distintas de 'fatal'
order_no_fatal = vehiculos_no_fatal.value_counts().index
sns.countplot(x=vehiculos_no_fatal, ax=axes[0], order=order_no_fatal,palette='colorblind')
axes[0].set_title('Vehículos involucrados en accidentes no fatales',fontsize=18)
axes[0].set_xlabel('')
axes[0].set_ylabel('Frecuencia')

# Agregar etiquetas de porcentaje en cada barra (Subgráfico 1)
total_no_fatal = len(vehiculos_no_fatal.dropna())
for p in axes[0].patches:
    percentage = '{:.1f}%'.format((p.get_height() / total_no_fatal) * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 0.1  # Ajuste para subir la etiqueta
    axes[0].annotate(percentage, (x, y), ha='center')

# Subgráfico 2: Gravedad 'fatal'
order_fatal = vehiculos_fatal.value_counts().index
sns.countplot(x=vehiculos_fatal, ax=axes[1], order=order_fatal,palette='colorblind')
axes[1].set_title('Vehículos involucrados en accidentes fatales',fontsize=18)
axes[1].set_xlabel('')
axes[1].set_ylabel('Frecuencia')

# Agregar etiquetas de porcentaje en cada barra (Subgráfico 2)
total_fatal = len(vehiculos_fatal)
for p in axes[1].patches:
    percentage = '{:.1f}%'.format((p.get_height() / total_fatal) * 100)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height() + 0.1  # Ajuste para subir la etiqueta
    axes[1].annotate(percentage, (x, y), ha='center')

# Ajustar el diseño para evitar solapamientos
plt.tight_layout()

# Mostrar los subgráficos
plt.show()

```

Conclusión: En los accidentes no fatales, se destaca que el vehículo más comúnmente involucrado es el automóvil, representando el 31% de los casos. Las motocicletas también son frecuentes, contribuyendo con un 28.2%, seguidas por los ciclistas con un 10.65%. En contraste, en accidentes fatales, las motocicletas están presentes en el 25.6% de los casos, seguidas por los automóviles con un 21.2%. Es relevante señalar que los peatones también están significativamente involucrados en accidentes fatales, representando un 19.2% de los casos.

```python
# Establecer estilo de Seaborn
sns.set(style="ticks")

# Filtrar gravedades distintas de 'fatal' y 'fatal'
no_fatales = hechos[hechos['gravedad'] != 'fatal']
fatales = hechos[hechos['gravedad'] == 'fatal']

# Crear subgráficos 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Subgráfico 1: Accidentes no fatales por año 
no_fatales_counts = no_fatales.groupby(pd.to_datetime(no_fatales['fecha']).dt.year).size()
sns.lineplot(x=no_fatales_counts.index, y=no_fatales_counts.values, ax=axes[0], marker='o', linewidth=1.5,markersize=10, linestyle='--')
axes[0].set_title('Accidentes no fatales por año',fontsize=18)
axes[0].set_xlabel('')
axes[0].set_ylabel('Número de accidentes')
sns.despine(ax=axes[0])

# Ajustar los ticks del eje x para mostrar solo años enteros
axes[0].set_xticks(no_fatales_counts.index)
axes[0].get_xaxis().get_major_formatter().set_useOffset(False)

# Subgráfico 2: Accidentes fatales por año
fatales_counts = fatales.groupby(pd.to_datetime(fatales['fecha']).dt.year).size()
sns.lineplot(x=fatales_counts.index, y=fatales_counts.values, ax=axes[1], marker='o', linewidth=1.5,markersize=10, linestyle='--')
axes[1].set_title('Accidentes fatales por año',fontsize=18)
axes[1].set_xlabel('')
axes[1].set_ylabel('Número de accidentes')
sns.despine(ax=axes[1])

# Ajustar los ticks del eje x para mostrar solo años enteros
axes[1].set_xticks(fatales_counts.index)
axes[1].get_xaxis().get_major_formatter().set_useOffset(False)

# Ajustar el diseño para evitar solapamientos
plt.tight_layout()

# Mostrar los subgráficos
plt.show()
```

Conclusión: Al analizar la evolución de los accidentes a lo largo de los años, se destaca una disminución significativa en la cantidad de accidentes no fatales desde el 2019 hasta el 2020. Sin embargo, esta tendencia positiva se revirtió en el 2021, donde los accidentes no fatales volvieron a aumentar. En contraste, los accidentes fatales experimentaron una reducción en el período del 2016 al 2017, seguido de un aumento en el 2018. Posteriormente, se registró una disminución notoria en los años 2019 y 2020, sugiriendo una posible influencia de las medidas de confinamiento relacionadas con la pandemia. Aunque los accidentes fatales aumentaron nuevamente en el 2021, no alcanzaron los niveles observados en 2018.

```python
# Convierte la columna 'fecha' a tipo datetime
hechos['fecha'] = pd.to_datetime(hechos['fecha'])

# Crea dos DataFrames separados para gravedad != 'fatal' y gravedad == 'fatal'
hechos_no_fatal = hechos[hechos['gravedad'] != 'fatal']
hechos_fatal = hechos[hechos['gravedad'] == 'fatal']

# Cuenta el número de accidentes por mes para ambos DataFrames
accidentes_por_mes_no_fatal = hechos_no_fatal.groupby(hechos_no_fatal['fecha'].dt.month)['id_hecho'].count()
accidentes_por_mes_fatal = hechos_fatal.groupby(hechos_fatal['fecha'].dt.month)['id_hecho'].count()

# Nombres de los meses
nombres_meses = [calendar.month_abbr[i] for i in range(1, 13)]

# Crea subplots con gráficos de líneas uno al lado del otro
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico de líneas para gravedad != 'fatal'
axs[0].plot(nombres_meses, accidentes_por_mes_no_fatal.values, marker='o', linestyle='--',markersize=10)
axs[0].set_title('Accidentes no fatales por mes', fontsize=18)
axs[0].set_ylabel('Número de Accidentes')
axs[0].set_ylim(1000, accidentes_por_mes_no_fatal.max()+300)  # Establece la escala desde 0 hasta el máximo valor

# Gráfico de líneas para gravedad == 'fatal'
axs[1].plot(nombres_meses, accidentes_por_mes_fatal.values, marker='o', linestyle='--',markersize=10)
axs[1].set_title('Accidentes fatales por mes', fontsize=18)
axs[1].set_xlabel('')
axs[1].set_ylim(20, accidentes_por_mes_fatal.max()+10)  # Establece la escala desde 0 hasta el máximo valor

plt.tight_layout()
plt.show()
```

Conclusión: En el caso de los accidentes no fatales, se observa que los meses con la mayor cantidad de incidentes fueron octubre, marzo y diciembre, mientras que los meses con la menor cantidad fueron mayo y abril.

En cuanto a los accidentes fatales, los meses que registraron la mayor cantidad de incidentes fueron diciembre, noviembre y agosto. En contraste, septiembre y octubre fueron los meses con la menor cantidad de accidentes fatales.
